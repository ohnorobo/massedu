```python
boundaries = (42.88679, 41.187053, -69.858861, -73.508142)
north, south, east, west = boundaries
b = Basemap(projection="lcc", 
            llcrnrlon=west,
            llcrnrlat=south,
            urcrnrlon=east,
            urcrnrlat=north,
            lat_0=(south+north)/2,
            lon_0=(east+west)/2)
unified_info = b.readshapefile("./shapefiles/tl_2013_25_unsd/tl_2013_25_unsd", "districts") #unified
unified_districts = b.districts
unified_districts_info = b.districts_info

secondary_info = b.readshapefile("./shapefiles/tl_2013_25_scsd/tl_2013_25_scsd", "districts") #secondary
secondary_districts = b.districts
secondary_districts_info = b.districts_info

districts = unified_districts + secondary_districts
districts_info = unified_districts_info + secondary_districts_info

fig = plt.figure(figsize=(20,10), dpi=10000)
#plt.title("All Districts")
```
